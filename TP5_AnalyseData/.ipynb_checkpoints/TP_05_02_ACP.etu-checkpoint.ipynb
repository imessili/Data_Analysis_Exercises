1) Chargez le dataset Iris et normalisez les données. Gardez les 'targets', c'est à dire les noms d'espèces de chaque plante dans un vecteur `y`. Vous pouvez utiliser l'option `return_X_y=True` de la fonction `load_iris`.


```python
from sklearn.datasets import load_iris
X,y = load_iris(return_X_y=True)
n,p= X.shape
```

```python
print(n,p)
```

```python
import numpy as np

mean_values = np.mean(X, axis=0)
std_dev_values = np.std(X, axis=0)

# Center the data
X_centered = X - mean_values

# Reduce the data
X_reduced = X_centered / std_dev_values

# The matrix X_r is now centered and reduced
Xr = X_reduced
```

2. Plottez les données `X` en 2D (les variables au choix) et affichez chaque point avec une couleur par espèce. Vous pouvez utiliser le paramètre `c=y` pour y arriver facilement.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Créer le plot en 2D
plt.figure(figsize=(8, 6))

# Tracer les points avec une couleur par espèce
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')

# Ajouter des titres et des étiquettes
plt.title('Visualisation des données iris en 2D')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Species')

# Afficher le plot
plt.grid(True)
plt.show()
```

3.Calculez la décomposition valeurs propres/vecteurs propres de la matrice de variance/covariance associées aux données normalisées.

```python
cov_matrix = np.dot(Xr.T, Xr) / (Xr.shape[0])
print(cov_matrix)
```

```python
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

# Afficher les valeurs propres et les vecteurs propres
print("Valeurs propres :\n", eigenvalues)
print("\nVecteurs propres :\n", eigenvectors)
```

4. Triez les valeurs propres dans l'ordre décroissant et appliquez le même ordre de tri aux colonnes de la matrice des vecteurs propres.

**Aide** : 

    1. On pourra utiliser la fonction `argsort` de `numpy`. [doc](https://numpy.org/doc/stable/reference/generated/numpy.argsort.html)
    2. L'utilisation des slices avec `-1` en dernière position permet de parcourir un élément de la fin au début. cf exemple.

```python
# Trouver l'indice de la plus grande valeur propre
max_eigenvalue_index = np.argmax(eigenvalues)

# Récupérer la plus grande valeur propre
max_eigenvalue = eigenvalues[max_eigenvalue_index]

# Récupérer le vecteur propre associé
max_eigenvector = eigenvectors[:, max_eigenvalue_index]

# Afficher la plus grande valeur propre et le vecteur propre associé
print("La plus grande valeur propre :", max_eigenvalue)
print("Le vecteur propre associé :", max_eigenvector)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Charger les données iris
X, y = load_iris(return_X_y=True)

# Effectuer l'ACP en 2D
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

# Matrice de projection (vecteurs propres)
P = pca.components_.T

# Projeter les données X avec P
X_projected = np.dot(X, P)

# Tracer les données en 2D avec une couleur par espèce
plt.figure(figsize=(8, 6))
plt.scatter(X_projected[:, 0], X_projected[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.title('Projection des données iris en 2D (ACP)')
plt.xlabel('Premier axe de l\'ACP')
plt.ylabel('Deuxième axe de l\'ACP')
plt.colorbar(label='Espèces')
plt.grid(True)
plt.show()
```

## Ex2 : Reconstruction
Ici nous allons étudier la reconstruction d'une image suite à une ACP, et après avoir gardé qu'une partie des composantes principales. Cet exercice nous permettra de mettre en évidence les capacités de compression des méthodes basées sur l'ACP.

1) Charger l'image du cameraman, une image bien connue pour le traitement d'images. Cette image étant en niveaux de gris, nous allons pouvoir la représenter sous forme d'une matrice `X` $\in \mathbb{R}^{256 \times 256}$. Ici, nous faisons l'hypothèse que chaque ligne de l'image est une observation, et que chaque colonne est une variable. Vous pouvez afficher l'image avec `plt.imshow(X,cmap="gray")`.

```python
!pip install imageio 
import imageio
import matplotlib.pyplot as plt
X = imageio.imread('TP_05_02_ACP_cameraman.png')
plt.imshow(X,cmap="gray");
```

```python
print(X.shape)
```

2. Calculez l'ACP (en utilisant votre code ou `sklearn`) de l'image codée par `X`

```python
import numpy as np
from sklearn.decomposition import PCA

# Effectuer l'ACP
pca = PCA()
X_pca = pca.fit_transform(X)
```

3. Calculez une projection de l'image en dimension 100 via votre ACP. Vous devriez pobtenir une matrice de taille $256 \times 100$

```python
# Perform PCA with 100 components
pca = PCA(n_components=100)
X_pca = pca.fit_transform(X)

# Reconstruct the image from the projection
X_reconstructed = pca.inverse_transform(X_pca)
print(X_pca.shape)

# Display the reconstructed image
plt.figure(figsize=(8, 6))
plt.imshow(X_pca, cmap="gray")
plt.title("Reconstructed Image (100 dimensions)")
plt.axis("off")
plt.show()

# Display the reconstructed image
plt.figure(figsize=(8, 6))
plt.imshow(X_reconstructed, cmap="gray")
plt.title("Reconstructed Image (100 dimensions)")
plt.axis("off")
plt.show()
```

4. Quelle est le pourcentage de variance expliquée par votre projection ? 

```python
# Calculer le pourcentage de variance expliquée
explained_variance_ratio = pca.explained_variance_ratio_

# Calculer le pourcentage total de variance expliquée
total_variance_explained = np.sum(explained_variance_ratio) * 100

print("Pourcentage total de variance expliquée :", total_variance_explained, "%")
```

```python
"""Le pourcentage de variance expliquée par une projection PCA
est une mesure importante qui indique la proportion de la variance
totale des données d'origine qui est conservée dans la projection PCA.
Cette mesure est cruciale pour évaluer à quel point la projection capture
efficacement la structure et les variations des données d'origine."""
```
