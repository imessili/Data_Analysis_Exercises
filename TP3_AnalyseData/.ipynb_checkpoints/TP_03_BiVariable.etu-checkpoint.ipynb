### Les variables $X$ et $Y$ sont elles indépendantes 
1. Calculez le tableau théorique sous l'hypothèse d'indépendance

```python
# Your contingency table
data1 = np.array([0.1, 0.2, 0.3])
data2 = np.array([0.2, 0.2, 0.0])

Tab = np.array([data1,
                data2])

# Calculate row and column sums
row_sums = np.sum(Tab, axis=1)
col_sums = np.sum(Tab, axis=0)

# Calculate total sum
total_sum = np.sum(Tab)

# Calculate expected frequencies
expected_frequencies = np.outer(row_sums, col_sums) / total_sum

print("Contingency Table:")
print(Tab)
print("\nMarginal Probabilities:")
print("Row Marginals:", row_sums)
print("Column Marginals:", col_sums)
print("\nExpected Frequencies (Under Independence):")
print(expected_frequencies)
```

2. Calculez la distance du $\chi_2$ à partir du tableau de contingence et du tableau théorique

```python
# Your contingency table
data1 = np.array([0.1, 0.2, 0.3])
data2 = np.array([0.2, 0.2, 0.0])

Tab = np.array([data1,
                data2])

# Calculate row and column sums
row_sums = np.sum(Tab, axis=1)
col_sums = np.sum(Tab, axis=0)

# Calculate total sum
total_sum = np.sum(Tab)

# Calculate expected frequencies
expected_frequencies = np.outer(row_sums, col_sums) / total_sum
print("Expected F (E):", expected_frequencies)

# Calculate Chi-squared distance
chi_squared_distance = np.sum((Tab - expected_frequencies)**2 / expected_frequencies)

print("Chi-squared distance:", chi_squared_distance)
```

```python
import csv
import numpy as np

#https://ourworldindata.org/co2-and-other-greenhouse-gas-emissions#co2-and-greenhouse-gas-emissions-country-profiles

tmp = []
min_year = 1900

with open('TP_03_BiVariable_co-emissions-per-capita.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if row['Entity'] == 'World':
            if int(int(row['Year']) >= min_year):
                tmp.append([float(row['Per capita CO2 emissions'])])
data_co2 = np.array(tmp)

tmp = []
with open('TP_03_BiVariable_temperature-anomaly.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if row['Entity'] == 'Global':
            if int(int(row['Year']) >= min_year):
                tmp.append([float(row['Median temperature anomaly from 1961-1990 average'])])
data_temp = np.array(tmp)

np.savez('TP_03_BiVariable_data_global_warming.npz', co2=data_co2, temp=data_temp)
```

3. Utilisez un bon indicateur quantitatif pour vous donner une idée si les deux variables sont dépendantes.

```python
# Convert NumPy arrays to pandas DataFrames
df_co2 = pd.DataFrame(data_co2, columns=["Per capita CO2 emissions"])
df_temp = pd.DataFrame(data_temp, columns=["Temperature Anomaly"])

# Concatenate the DataFrames horizontally
df_combined = pd.concat([df_co2, df_temp], axis=1)

# Display the combined DataFrame
display(df_combined)
```

```python
## A correlation coefficient of 0.789 indicates a strong positive correlation between the two variables 
correlation_matrix = df_combined.corr()

correlation_coefficient = correlation_matrix.loc['Per capita CO2 emissions', 'Temperature Anomaly']

print(f'Correlation coefficient: {correlation_coefficient}')
```

```python
import numpy as np

# Assuming data_co2 and data_temp are the columns from your DataFrame
data_co2 = df_combined['Per capita CO2 emissions'].values
data_temp = df_combined['Temperature Anomaly'].values

# Step 1: Standardize the data
std_data_co2 = (data_co2 - np.mean(data_co2)) / np.std(data_co2)
std_data_temp = (data_temp - np.mean(data_temp)) / np.std(data_temp)

# Step 2: Calculate the covariance
covariance = np.cov(std_data_co2, std_data_temp, ddof=0)[0, 1]

# Step 3: Calculate the correlation coefficient
correlation_coefficient = covariance / (np.std(std_data_co2) * np.std(std_data_temp))

print(f'Correlation coefficient (using NumPy): {correlation_coefficient}')
```

```python
#This is also called persons R
# Extract the columns
co2_data = df_combined['Per capita CO2 emissions']
temp_data = df_combined['Temperature Anomaly']

# Calculate means
mean_co2 = co2_data.mean()
mean_temp = temp_data.mean()

# Calculate standard deviations
std_co2 = co2_data.std()
std_temp = temp_data.std()

# Calculate the correlation coefficient
r = np.sum((co2_data - mean_co2) * (temp_data - mean_temp)) / (len(df_combined) * std_co2 * std_temp)

print(f'Correlation coefficient: {r}')
```

3\. Construisez le tableau de contingence $O$ entre les variables $x$ et $g$

```python
Un tableau de contingence (aussi appelé tableau croisé) est une manière de représenter la distribution
conjointe de deux variables catégorielles. Il est souvent utilisé en statistique descriptive et en
analyse de données pour examiner les relations entre deux variables qualitatives.

Ce tableau présente le nombre d'occurrences de chaque paire de catégories pour les deux variables.

Un tableau de contingence est généralement disposé sous la forme d'un tableau à deux entrées,
où les lignes représentent les modalités d'une variable et les colonnes représentent les modalités de
l'autre variable. Les cellules du tableau contiennent les fréquences ou les comptages observés pour
chaque paire de catégories.
```

```python
# Récupérer les deux premières lignes de la matrice (école et genre)
ecole = student_data[:, 0]
genre = student_data[:, 1]

# Calculer le tableau de contingence
tableau_contingence = np.zeros((3, 3), dtype=int)

for i in range(len(ecole)):
    tableau_contingence[ecole[i], genre[i]] += 1

## Calcul des marges
tableau_contingence[:, -1] = np.sum(tableau_contingence, axis=1)
tableau_contingence[-1, :] = np.sum(tableau_contingence, axis=0)

# Afficher le tableau de contingence
print("Tableau de contingence :")
print(tableau_contingence)
```

4\. Construisez le tableau théorique associé en supposant l’indépendance des deux variables.

```python
tableau_contingence_theo = np.zeros((2, 2))

for i in range(2):
    for j in range(2):
        tableau_contingence_theo[i,j] = (tableau_contingence[i,2]/tableau_contingence[2,2]) * (tableau_contingence[2,j]/tableau_contingence[2,2])

print(tableau_contingence_theo)
```

5\. Calculez la distance du chi2 entre les variables x et g. Ces deux variables sont elles liées ou sont elles indépendantes ?

```python
# Récupérer les deux premières lignes de la matrice (école et genre)
ecole = student_data[:, 0]
genre = student_data[:, 1]

# Calculer le tableau de contingence
tableau = np.zeros((2, 2), dtype=int)

for i in range(len(ecole)):
    tableau[ecole[i], genre[i]] += 1
    
# Calculate Chi-squared distance
chi_squared_distance = np.sum((tableau - tableau_contingence_theo)**2 / tableau_contingence_theo)

print("Chi-squared distance:", chi_squared_distance)
```
