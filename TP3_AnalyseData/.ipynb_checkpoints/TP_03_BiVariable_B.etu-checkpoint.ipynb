```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.path import Path
import scipy
import scipy.spatial as scsp
import sklearn
import sklearn.datasets
```

Compléter la classe suivante qui permet de faire une représentation en Oignon

```python
## Compléter la classe suivante

class Onion:

    @classmethod
    def _convexhullvertices(cls,data):
        """
        Compute the convex hull of points in data and
        return the indices of the corresponding points.
        
        Parameters
        ----------
        data : ndarray of double, shape (npoints, ndim)
            Coordinates of input points.

        Returns
        -------
        ndarray of ints, shape (nvertices,)
            Indices of points forming the vertices of the convex hull.
            For 2-D convex hulls, the vertices are in counterclockwise order.
            For other dimensions, they are in input order.
        """
        hull = scsp.ConvexHull(data)
        return hull.vertices


    @classmethod
    def plot_hull(cls, X, ax=None):
        """
        Plot a hull
        
        Parameters
        ----------
        X : ndarray of double, shape (npoints, ndim)
            Coordinates of hull points.
        ax : plot axis
        """
        axx = plt if ax is None else ax
        if X.shape[0]>2:
            Xc = np.concatenate((X,X[0,np.newaxis]),axis=0)
        else:
            Xc = X
        axx.plot(Xc[:,0],Xc[:,1]) 


    @classmethod
    def _fit(cls,X):
        
        
        print(X)
        
        """
        Compute reccursively the onion of the data X

        Parameters
        ----------
        X : ndarray of double, shape (npoints, ndim)
            Coordinates of input points.

        Returns
        -------
        ndarray of double, shape (npoints, ndim)
            Sorted X with the inner most leaf at first
        ndarrays of ints (nleafs,)
            A list of first index in axis 0 of Sorted X of each leaf
            Always starts by 0
        """

        raise NotImplementedError
   

    def __init__(self):
        """
        Initializes attributes of the Onion class
        """
        # Affected when fitted

        ## ndarray of double, shape (npoints, ndim)
        ## Sorted X with the inner most leaf at first
        self.data = None

        ## ndarrays of ints (nleafs,)
        ## A list of first index of each leaf in axis 0 of sorted X 
        self.bulb = None 

        # Cached computation

        ## mean of the inner most leaf
        ## ndarray of double, shape (ndim,)
        self.median = None
        
        ## ndarray of double, shape (n, ndim)
        ## leaf containing inside its hull half of the points
        self.median_bag = None

        ## ndarray of double, shape (n, ndim)
        ## median_bag inflated 3 times from the median
        self.sketch = None

        ## ndarrays of ints
        ## list of points inside the sketch
        self.inliers = None

        ## ndarrays of ints
        ## list of points outside the sketch
        self.outliers = None

        ## ndarray of double, shape (n, ndim)
        ## convex hull of inliers points
        self.whiskers = None

    def fit(self,X):
        """
        Compute the ognon of the data X

        Modified Attributes
        -------------------
        all set to None except:

        data : ndarray of double, shape (npoints, ndim)
        bulb : list of ndarrays of ints


        Parameters
        ----------
        X : ndarray of double, shape (npoints, ndim)
            Coordinates of input points.

        """
        self.__init__() # We erase all
        self.data, self.bulb = self._fit(X)

    def plot_bulb(self, ax=None):
        """
        Plot the bulb after being fitted
        
        Parameters
        ----------
        ax : plot axis
        """
        assert self.data is not None # we have fitted
        assert self.data.shape[1] == 2 # we are in 2D

        axx = plt if ax is None else ax

        bulb_start = self.bulb
        bulb_end = list(self.bulb[1:])+[self.data.shape[0]]
        
        for s,e in zip(bulb_start,bulb_end):
            raise NotImplementedError
    
    def get_median(self):
        """
        Compute and return the median of the bulb.
        
        Modified Attributes
        -------------------

        median: ndarray of double, shape (ndim,)
            The median
        Returns
        -------
        ndarray of double, shape (ndim,)
            The median
        """
        assert self.data is not None
        if self.median is not None:
            return self.median

        inner_leaf = self.data[:self.bulb[1]]
        #self.median = 
        raise NotImplementedError
        
        return self.median
    
    def get_median_bag(self):
        """
        Compute and return the median bag of the bulb;
        i.e. the smallest leaf countaining half of the points.
        
        Modified Attributes
        -------------------

        median_bag: ndarray of double, shape (n, ndim)
            The points of the hull of the median bag leaf
        
        Returns
        -------
        ndarray of double, shape (n, ndim)
            The points of the hull of the median bag leaf
        """
        assert self.data is not None
        if self.median_bag is not None:
            return self.median_bag

        bulb_start = self.bulb
        bulb_end = list(self.bulb[1:])+[self.data.shape[0]]
        for s,e in zip(bulb_start,bulb_end):
            raise NotImplementedError
        else:
            self.median_bag = self.data[bulb_start[-1],bulb_end[-1]]

        return self.median_bag

    def get_sketch(self):
        """
        Compute and return the sketch of the bulb;
        i.e. the median bag inflated 3 times from the median.

        Modified Attributes
        -------------------

        sketch: ndarray of double, shape (n, ndim)
            The points of the sketch
        
        Returns
        -------
        ndarray of double, shape (n, ndim)
            The points of the sketch
        """
        assert self.data is not None
        if self.sketch is not None:
            return self.sketch

        self.sketch = (self.get_median_bag()-self.get_median()[np.newaxis,:])*3+self.get_median()[np.newaxis,:]

        return self.sketch
    

    def _liers(self):
        """
        Compute the points inside and outside the sketch
        
        Modified Attributes
        -------------------
        inliers : ndarrays of ints, list of points inside the sketch
        outliers : ndarrays of ints, list of points outside the sketch
        """
        assert self.data is not None # we have fitted
        assert self.data.shape[1] == 2 # we are in 2D

        sketchPath = Path(self.get_sketch())
        self.inliers, = np.nonzero([sketchPath.contains_point(x) for x in self.data])
        self.outliers, = np.nonzero([not(sketchPath.contains_point(x)) for x in self.data])       

    def get_inliers(self):
        """
        Compute and return the points inside the sketch
        
        Modified Attributes
        -------------------
        inliers : ndarrays of ints, list of points inside the sketch

        Returns
        -------
        ndarrays of ints
            list of points inside the sketch
        """
        assert self.data is not None # we have fitted
        assert self.data.shape[1] == 2 # we are in 2D

        if self.inliers is not None:
            return self.inliers
        
        self._liers()

        return self.inliers
    
    def get_outliers(self):
        """
        Compute and return the points outside the sketch
        
        Modified Attributes
        -------------------
        outliers : ndarrays of ints, list of points outside the sketch 

        Returns
        -------
        ndarrays of ints
            list of points outside the sketch
        """
        assert self.data is not None # we have fitted
        assert self.data.shape[1] == 2 # we are in 2D

        if self.outliers is not None:
            return self.outliers
        
        self._liers()

        return self.outliers
    
    def get_whiskers(self):
        """
        Compute and return the hull of the points inside the sketch
        
        Modified Attributes
        -------------------
        whiskers : ndarray of double, shape (n, ndim),
            convex hull of inliers points

        Returns
        -------
        ndarray of double, shape (n, ndim)
            convex hull of inliers points
        """
        assert self.data is not None # we have fitted
        assert self.data.shape[1] == 2 # we are in 2D

        if self.whiskers is not None:
            return self.whiskers
        
        X = self.data[self.get_inliers()]
        border = self._convexhullvertices(X)
        self.whiskers = X[border]

        return self.whiskers
    
    def plot_boxplot(self, ax=None):
        """
        Plot the boxplot after being fitted
        
        Parameters
        ----------
        ax : plot axis
        """
        assert self.data is not None # we have fitted
        assert self.data.shape[1] == 2 # we are in 2D

        axx = plt if ax is None else ax

        indata = self.data[self.get_inliers()]
        outdata = self.data[self.get_outliers()]

        axx.plot(indata[:,0],indata[:,1],'b+')
        axx.plot(outdata[:,0],outdata[:,1],'b*')
        median = self.get_median()
        axx.plot(median[0],median[1],'ro')
        self.plot_hull(self.get_median_bag(),ax=axx)
        self.plot_hull(self.get_whiskers(),ax=axx)

    @classmethod
    def boxplot(cls,X, ax=None):
        """
        Plot the boxplot of data X
        
        Parameters
        ----------
        ax : plot axis
        """
        obj = cls()
        obj.fit(X)
        obj.plot_boxplot(ax=ax)
```

```python
# Dataset artificiel

np.random.seed(42)
X,y = sklearn.datasets.make_moons(n_samples=100,noise=0.1)
data = X[np.nonzero(y==1)]
plt.figure()
plt.plot(data[:,0],data[:,1],'+')

# On rajoute deux outliers
data = np.concatenate((data,np.array([[-2,-.5],[3,-.5]])),axis=0)

plt.figure()
plt.plot(data[:,0],data[:,1],'+')
```

```python
## Test de la classe

anOnion=Onion()
anOnion.fit(data)

plt.figure()
plt.plot(data[:,0],data[:,1],'+')
anOnion.plot_bulb()

plt.figure()
plt.plot(data[:,0],data[:,1],'+')
median = anOnion.get_median()
plt.plot(median[0],median[1],'ro')
anOnion.plot_hull(anOnion.get_median_bag())
anOnion.plot_hull(anOnion.get_sketch())
anOnion.plot_hull(anOnion.get_whiskers())

plt.figure()
anOnion.plot_boxplot()

plt.figure()
Onion.boxplot(data)
```
